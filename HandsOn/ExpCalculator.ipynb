# Computing the QFT Output by Hand

This is a **purely classical** scratch-pad notebook. There is no Qiskit code to update – we just want to verify, with numpy, the interference pattern that the Quantum Fourier Transform (QFT) part of Shor's algorithm produces when the period of the function is small.

Concretely, after the QFT is applied to the phase register, the amplitude of measuring outcome `y` is proportional to

\[\sum_{k} e^{-2\pi i \cdot \frac{y \cdot x_k}{N}}\]

over the `x_k` values that were "marked" by the modular-exponentiation oracle. We hard-code those values for a 4-qubit (`N = 16`) example below.

```python
import numpy as np
```

## The interference pattern

For each integer `y` (the value the quantum circuit could measure on its phase register), we sum the four complex exponentials. Whenever the sum is **non-zero** the corresponding `y` is an outcome the QFT will produce with non-zero probability.

You'll see that only a handful of `y` values survive – exactly the ones whose `y/N` (with `N=16`) is an approximation of the underlying *period*.

```python
for y in range(15):
    result = np.exp(-1j*3*np.pi/8*y) + np.exp(-1j*7*np.pi/8*y) + np.exp(-1j*11*np.pi/8*y) + np.exp(-1j*15*np.pi/8*y)
    if abs(result) < 1e-10: 
        result = 0
    print(y, result)
```
