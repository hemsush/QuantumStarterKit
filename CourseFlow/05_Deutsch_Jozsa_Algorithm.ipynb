## Algorithm implementation

For a constant oracle the input register measures all zero; a balanced oracle produces a non-zero bit string.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
def dj_circuit(n,kind):
    qc=QuantumCircuit(n+1,n); qc.x(n); qc.h(range(n+1)); qc.barrier()
    if kind=="constant1": qc.x(n)
    elif kind=="balanced":
        for q in range(n): qc.cx(q,n)
    qc.barrier(); qc.h(range(n)); qc.measure(range(n),range(n)); return qc
sim=AerSimulator()
for kind in ["constant0","constant1","balanced"]:
    qc=dj_circuit(3,kind); counts=sim.run(qc,shots=1024).result().get_counts()
    print(kind,counts); display(qc.draw("mpl")); display(plot_histogram(counts))
```

## Automatic classification

The decision rule depends only on whether the measured string is all zeros.

```python
for kind in ["constant0","balanced"]:
    counts=sim.run(dj_circuit(3,kind),shots=1).result().get_counts()
    bits=next(iter(counts)); print(kind,"classified as", "constant" if set(bits)=={"0"} else "balanced")
```
