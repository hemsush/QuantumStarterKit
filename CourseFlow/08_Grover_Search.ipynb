## Two-qubit Grover search

This oracle marks $|11
angle$ by phase; the diffuser reflects amplitudes about their mean.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
qc=QuantumCircuit(2,2); qc.h([0,1]); qc.cz(0,1)
qc.h([0,1]); qc.x([0,1]); qc.h(1); qc.cx(0,1); qc.h(1); qc.x([0,1]); qc.h([0,1])
qc.measure([0,1],[0,1])
counts=AerSimulator().run(qc,shots=1024).result().get_counts()
display(qc.draw("mpl")); print(counts); display(plot_histogram(counts))
```

## Mark another state

X gates around the $|11
angle$ oracle can mark a different basis state.

```python
def oracle_for(bits):
    o=QuantumCircuit(2,name=f"mark {bits}")
    for q,b in enumerate(reversed(bits)):
        if b=="0": o.x(q)
    o.cz(0,1)
    for q,b in enumerate(reversed(bits)):
        if b=="0": o.x(q)
    return o
display(oracle_for("01").draw("mpl"))
```
