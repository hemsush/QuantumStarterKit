## Hidden-string oracle

The oracle computes $f(x)=s\cdot x\pmod2$.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
def bv(secret):
    n=len(secret); qc=QuantumCircuit(n+1,n); qc.x(n); qc.h(range(n+1)); qc.barrier()
    for q,bit in enumerate(reversed(secret)):
        if bit=="1": qc.cx(q,n)
    qc.barrier(); qc.h(range(n)); qc.measure(range(n),range(n)); return qc
secret="101101"; qc=bv(secret)
counts=AerSimulator().run(qc,shots=256).result().get_counts()
display(qc.draw("mpl")); print("Secret:",secret,"Recovered:",max(counts,key=counts.get)); display(plot_histogram(counts))
```

## Multiple tests

Confirm the implementation for several strings.

```python
for s in ["0000","0001","1010","1111"]:
    c=AerSimulator().run(bv(s),shots=32).result().get_counts(); print(s,max(c,key=c.get))
```
