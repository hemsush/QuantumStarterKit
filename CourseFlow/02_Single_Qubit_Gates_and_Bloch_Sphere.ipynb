## Reusable visualizer

Each circuit is converted directly to a statevector before measurement.

```python
import numpy as np
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator
from qiskit.visualization import plot_bloch_multivector
def inspect(qc):
    state=Statevector.from_instruction(qc)
    display(qc.draw("mpl")); display(state.draw("latex")); display(plot_bloch_multivector(state))
    return state
```

## Pauli X and Y

X flips the basis state; Y flips it and adds phase.

```python
for gate in ["x","y"]:
    qc=QuantumCircuit(1); getattr(qc,gate)(0)
    print(gate.upper(), np.asarray(Operator(qc))); inspect(qc)
```

## Pauli Z and Hadamard

Z is visible on a superposition: $Z|+
angle=|−
angle$.

```python
for gate in ["h","z"]:
    qc=QuantumCircuit(1); qc.h(0)
    if gate=="z": qc.z(0)
    print("H" if gate=="h" else "HZ"); inspect(qc)
```

## Phase and rotation gates

Prepare $|+
angle$ so phase rotations appear on the Bloch sphere.

```python
circuits={}
for name in ["s","t"]:
    qc=QuantumCircuit(1); qc.h(0); getattr(qc,name)(0); circuits[name]=qc
qc=QuantumCircuit(1); qc.h(0); qc.rz(np.pi/2,0); circuits["rz(pi/2)"]=qc
for name,qc in circuits.items(): print(name); inspect(qc)
```

## Gate identities

Verify $X^2=H^2=I$ using operator equivalence.

```python
for gate in ["x","h"]:
    qc=QuantumCircuit(1); getattr(qc,gate)(0); getattr(qc,gate)(0)
    print(gate, Operator(qc).equiv(Operator(np.eye(2))))
```
