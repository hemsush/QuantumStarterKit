## CNOT truth table

Qiskit labels basis states as $|q_1q_0
angle$.

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector, Operator
for q1,q0 in [(0,0),(0,1),(1,0),(1,1)]:
    qc=QuantumCircuit(2)
    if q0: qc.x(0)
    if q1: qc.x(1)
    qc.cx(0,1)
    print(f"input |{q1}{q0}> ->", Statevector.from_instruction(qc).probabilities_dict())
```

## Bell state

H followed by CNOT prepares $(|00\rangle+|11\rangle)/\sqrt2$.

```python
from qiskit.visualization import plot_state_city, plot_histogram
bell=QuantumCircuit(2); bell.h(0); bell.cx(0,1)
state=Statevector.from_instruction(bell)
display(bell.draw("mpl")); display(state.draw("latex")); display(plot_state_city(state))
```

## Bell correlations

Individual results are random, but the two bits agree.

```python
from qiskit_aer import AerSimulator
measured=bell.copy(); measured.measure_all()
counts=AerSimulator().run(measured,shots=2048).result().get_counts()
print(counts); display(plot_histogram(counts))
```

## CZ and CNOT equivalence

Hadamards on the CZ target convert CZ into CX.

```python
cz_as_cx=QuantumCircuit(2); cz_as_cx.h(1); cz_as_cx.cz(0,1); cz_as_cx.h(1)
cx=QuantumCircuit(2); cx.cx(0,1)
print("Equivalent:",Operator(cz_as_cx).equiv(Operator(cx))); display(cz_as_cx.draw("mpl"))
```

## SWAP and Toffoli

SWAP equals three CNOTs; Toffoli writes classical AND to an initially zero target.

```python
swap=QuantumCircuit(2); swap.swap(0,1)
three_cx=QuantumCircuit(2); three_cx.cx(0,1); three_cx.cx(1,0); three_cx.cx(0,1)
print("SWAP equivalence:",Operator(swap).equiv(Operator(three_cx)))
tof=QuantumCircuit(3); tof.x([0,1]); tof.ccx(0,1,2)
display(tof.draw("mpl")); print(Statevector.from_instruction(tof).probabilities_dict())
```
