## Basis states

A qubit is $|\psi\rangle=\alpha|0\rangle+\beta|1\rangle$ with $|\alpha|^2+|\beta|^2=1$.

```python
import numpy as np
from qiskit.quantum_info import Statevector
zero, one = Statevector.from_label("0"), Statevector.from_label("1")
display(zero.draw("latex")); display(one.draw("latex"))
print("Probabilities |0>:", zero.probabilities_dict())
```

## Custom normalized state

The squared magnitude of each amplitude is its measurement probability.

```python
psi = Statevector([np.sqrt(0.7), 1j*np.sqrt(0.3)])
print("Valid state:", psi.is_valid())
display(psi.draw("latex")); print(psi.probabilities_dict())
```

## Superposition and measurement

Hadamard prepares $|+
angle$. Increasing shots makes observed frequencies approach 50:50.

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
qc=QuantumCircuit(1,1); qc.h(0); qc.measure(0,0)
sim=AerSimulator()
all_counts={n:sim.run(qc,shots=n).result().get_counts() for n in [100,1000,10000]}
display(qc.draw("mpl")); print(all_counts)
display(plot_histogram(list(all_counts.values()),legend=[f"{n} shots" for n in all_counts]))
```
